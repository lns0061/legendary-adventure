```python

import numpy as np
import matplotlib.pyplot as plt

r1 = 47
l = 5.6*10**(-3)
w = 10**4
c = 50*10**(-9)
x = np.arange(7500, 12001, 100)
result = r1/((r1**2 + (2*np.pi*x*l - 1/(2*np.pi*x*c))**2)**(1/2))

plt.plot(x, result)
plt.xlabel('frecuency')
plt.ylabel('k')
plt.grid(True)
plt.show()
plt.plot(x, np.arctan(-(2*np.pi*x*l-1/(2*np.pi*x*c))/r1))
plt.xlabel('frecuency')
plt.ylabel('arctg(phi)')
plt.grid(True)
plt.show()
```
