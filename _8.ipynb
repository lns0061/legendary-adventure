```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Параметры детектора и источника
R = 5.0  # Радиус детектора в см
D = 10.0  # Высота детектора в см
X0, Y0, Z0 = 2.0, 2.0, 3.0  # Координаты источника

# Параметры источника Cs-137
energy_peaks = [0.6617]  # Основной пик энергии в МэВ для Cs-137

# Количество энергетических групп
energy_groups = 1024
max_energy = 1.0  # Максимальная энергия в МэВ для моделирования
energy_bins = np.linspace(0, max_energy, energy_groups)

# Моделирование регистрации частиц
def simulate_detector_response(R, D, X0, Y0, Z0, energy_peaks, energy_bins):
    spectrum = np.zeros(len(energy_bins))
    for peak in energy_peaks:
        counts = norm.pdf(energy_bins, loc=peak, scale=0.01)  # Размытие пика
        spectrum += counts
    return spectrum

# Получение спектра
spectrum = simulate_detector_response(R, D, X0, Y0, Z0, energy_peaks, energy_bins)

# Визуализация спектра
plt.plot(energy_bins, spectrum, label='Cs-137')
plt.xlabel('Энергия (МэВ)')
plt.ylabel('Количество зарегистрированных частиц')
plt.title('Спектр регистрации гамма-излучения Cs-137')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Параметры детектора и источника
R = 5.0  # Радиус детектора в см
D = 10.0  # Высота детектора в см
X0, Y0, Z0 = 2.0, 2.0, 3.0  # Координаты источника

# Параметры источника Co-60
energy_peaks_Co60 = [1.173, 1.332]  # Основной пик энергии в МэВ для Co-60, можно менять для любых изототов !!!

# Количество энергетических групп
energy_groups = 1024
max_energy = 2.0  # Максимальная энергия в МэВ для моделирования, так же можно менять для любых изотопов(главное чтобы энергетические пики входили в интервал)
energy_bins = np.linspace(0, max_energy, energy_groups)

# Моделирование регистрации частиц
def simulate_detector_response(R, D, X0, Y0, Z0, energy_peaks, energy_bins):
    spectrum = np.zeros(len(energy_bins))
    for peak in energy_peaks:
        counts = norm.pdf(energy_bins, loc=peak, scale=0.01)  # Размытие пика, погрешность измерения, в данном случае грубо 3 сигма
        spectrum += counts
    return spectrum
spectrum_Co60 = simulate_detector_response(R, D, X0, Y0, Z0, energy_peaks_Co60, energy_bins)

plt.plot(energy_bins, spectrum_Co60, label='Co-60')
plt.xlabel('Энергия (МэВ)')
plt.ylabel('Количество зарегистрированных частиц')
plt.title('Спектр регистрации гамма-излучения Co-60')
plt.legend()
plt.show()
```
