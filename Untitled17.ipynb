```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
N = 1000
azimuthal_angles = np.random.uniform(0, 2*np.pi, N)
polar_angles = np.arccos(2*np.random.uniform(0, 1, N) - 1)

r = 1.0
x = r * np.sin(polar_angles) * np.cos(azimuthal_angles)
y = r * np.sin(polar_angles) * np.sin(azimuthal_angles)
z = r * np.cos(polar_angles)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('        ')
plt.show()
```
