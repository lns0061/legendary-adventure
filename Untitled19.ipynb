```python
import matplotlib.pyplot as plt
import numpy as np

def plot_energy_lines(num_energies):
    energies = np.random.uniform(0, 100, num_energies)

    plt.figure()
    plt.plot(range(1, num_energies+1), energies, marker='o', linestyle='-', color='b')
    plt.xlabel('Линии энергии')
    plt.ylabel('Значение энергии')
    plt.xlabel('Номер линии')
    plt.ylabel('Энергия, МэВ')
    plt.title(f'График для {num_energies} энергетических линий')
    plt.grid(True)
    plt.show()

plot_energy_lines(5)  # График для 5 энергетических линий
plot_energy_lines(10)  # График для 10 энергетических линий
plot_energy_lines(100)  # График для 100 энергетических линий
```
