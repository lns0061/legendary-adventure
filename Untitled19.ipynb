```python
import matplotlib.pyplot as plt
import numpy as np

def plot_energy_lines(num_energies):
    energies = np.random.uniform(0, 100, num_energies)

    plt.figure()
    plt.plot(range(1, num_energies+1), energies, marker='o', linestyle='-', color='b')
    plt.xlabel(' ')
    plt.ylabel(' ')
    plt.xlabel(' ')
    plt.ylabel(', ')
    plt.title(f'  {num_energies}  ')
    plt.grid(True)
    plt.show()

plot_energy_lines(5)  #   5  
plot_energy_lines(10)  #   10  
plot_energy_lines(100)  #   100  
```
